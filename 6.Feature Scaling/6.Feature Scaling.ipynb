```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset =  pd.read_csv(r"C:\Users\Admin\Downloads\loan.csv")
```

```python
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset["Loan_Amount_Term"].fillna(dataset["Loan_Amount_Term"].mean(),inplace=True)
```

```python
dataset.fillna({"Loan_Amount_Term":dataset["Loan_Amount_Term"].mean()},inplace=True)
```

```python
sns.distplot(dataset["Loan_Amount_Term"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
af = StandardScaler()
af.fit(dataset[["Loan_Amount_Term"]])
```

```python
af.transform(dataset[["Loan_Amount_Term"]])
```

```python
dataset["Loan_Amount_Term_ss"] = pd.DataFrame(af.transform(dataset[["Loan_Amount_Term"]]), columns=["x"])
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["Loan_Amount_Term"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["Loan_Amount_Term_ss"])

plt.show()
```

# Feature Scaling (Nomramlization)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
dataset.head()
```

```python
mm = MinMaxScaler()
mm.fit(dataset[["LoanAmount"]])
```

```python
dataset["LoanAmount_min"] = mm.transform(dataset[["LoanAmount"]])
```

```python
dataset.head()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["LoanAmount"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["LoanAmount_min"])

plt.show()
```
