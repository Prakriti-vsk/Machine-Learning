# Outlier Removal Using IQR

```python
dataset.shape
```

```python
q1 = dataset['LoanAmount'].quantile(0.25)
q3 = dataset['LoanAmount'].quantile(0.75)
q3
```

```python
IQR = q3-q1
```

```python
IQR
```

```python
min_range = q1 - (1.5*IQR)
max_range = q3 + (1.5*IQR)
min_range, max_range
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x="LoanAmount", data=dataset)
plt.show()
```

```python
dataset
```

```python
dataset[dataset["LoanAmount"]<=max_range]
```

```python
dataset[dataset["LoanAmount"]<=max_range]
```

```python
new_dataset = dataset[dataset["LoanAmount"]<=max_range]
```

```python
new_dataset.shape
```

```python
sns.boxplot(x="LoanAmount",data=new_dataset)
plt.show()
```

```python
min_range = dataset1["LoanAmount"].mean() - (3*dataset["LoanAmount"].std())
max_range = dataset1["LoanAmount"].mean() + (3*dataset["LoanAmount"].std())
```

```python
min_range,max_range
```

```python
dataset1[dataset1["LoanAmount"]<=max_range]
```

```python
new_dataset1 = dataset1[dataset1["LoanAmount"]<=max_range]
```

```python
dataset1["z_score"]=z_score
```
