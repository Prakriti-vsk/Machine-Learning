### **Handling Missing Values (Imputing Categorical Data)**

```python
df1 =  pd.read_csv(r"C:\Users\Admin\Downloads\loan.csv")
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1.fillna(10)
```

```python
df1.bfill()
```

```python
df1.ffill()
```

```python
df1.bfill(axis=1)
```

```python
df1.ffill(axis=1)
```

```python
df1.isnull().sum()
```

```python
df1["Gender"].mode()
```

```python
df1.fillna({"Gender":df1["Gender"].mode()},inplace=True)
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
df1.select_dtypes(include="object").columns
```

```python
for i in df1.select_dtypes(include="object").columns:
    df1.fillna({i:df1[i].mode()},inplace=True)
```

```python
df1.isnull().sum()
```

```python
df1.select_dtypes(include="object").columns
```

```python
df1.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
ni = SimpleImputer(strategy='mean')
nr = ni.fit_transform(df1[['LoanAmount','Loan_Amount_Term','Credit_History']])
```

```python
new_df = pd.DataFrame(nr,columns=df1.select_dtypes(include="float64").columns)
```
